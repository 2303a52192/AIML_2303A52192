LInear Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("/content/hw_200.csv")
df.head()
```

```python
df.columns = ['Index', 'Height', 'Weight']
data = df.drop(columns=['Index'])
```

```python
X = data['Height'].values.reshape(-1, 1)
y = data['Weight'].values
```

```python
simple_model = LinearRegression()
simple_model.fit(X, y)
y_pred = simple_model.predict(X)
```

```python
r2_simple = r2_score(y, y_pred)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='blue', alpha=0.6, label='Actual data')
plt.plot(X, y_pred, color='red', label='Regression line')
plt.title('Simple Linear Regression: Height vs Weight')
plt.xlabel('Height (Inches)')
plt.ylabel('Weight (Pounds)')
plt.legend()
plt.grid()
plt.show()
```

```python
print(f"R-squared (Simple Linear Regression): {r2_simple:.2f}")
```

```python
data['Height_squared'] = data['Height'] ** 2
X_multi = data[['Height', 'Height_squared']]
y_multi = data['Weight']
```

```python
multi_model = LinearRegression()
multi_model.fit(X_multi, y_multi)
y_multi_pred = multi_model.predict(X_multi)
```

```python
r2_multi = r2_score(y_multi, y_multi_pred)
```

```python
print(f"R-squared (Multiple Linear Regression): {r2_multi:.2f}")
```

```python
residuals = y_multi - y_multi_pred
plt.figure(figsize=(8, 6))
plt.scatter(y_multi_pred, residuals, color='green', alpha=0.6)
plt.axhline(0, color='red', linestyle='--')
plt.title('Residual Plot: Multiple Linear Regression')
plt.xlabel('Predicted Weight')
plt.ylabel('Residuals')
plt.grid()
plt.show()
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_multi_scaled = scaler.fit_transform(X_multi)
```

```python
scaled_simple_model = LinearRegression()
scaled_simple_model.fit(X_scaled, y)
y_scaled_pred = scaled_simple_model.predict(X_scaled)
r2_scaled_simple = r2_score(y, y_scaled_pred)
```

```python
scaled_multi_model = LinearRegression()
scaled_multi_model.fit(X_multi_scaled, y_multi)
y_scaled_multi_pred = scaled_multi_model.predict(X_multi_scaled)
r2_scaled_multi = r2_score(y_multi, y_scaled_multi_pred)

```

```python
print(f"R-squared (Simple Regression with Scaling): {r2_scaled_simple:.2f}")
print(f"R-squared (Multiple Regression with Scaling): {r2_scaled_multi:.2f}")
```
